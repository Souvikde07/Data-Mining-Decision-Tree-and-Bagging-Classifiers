```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn as skl
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier 
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_validate
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, make_scorer
import mlxtend as ml
from mlxtend.plotting import plot_learning_curves
```

```python
dd = pd.read_csv('SUSY.csv',header=None)
dd.head()
```

```python
x = dd.iloc[:,1:]
y = dd[0]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, shuffle=True, random_state=101)
```

```python
print(x_train.shape, y_train.shape, x_test.shape, y_test.shape)
```

```python
classifier = DecisionTreeClassifier(criterion="gini", random_state=1)
```

```python
classifier.fit(x_train,y_train)
```

```python
classifier.score(x_test,y_test)
```

```python
y_predict = classifier.predict(x_test)
```

```python
plot_learning_curves(x_train, y_train, x_test, y_test, classifier, scoring="misclassification error")
```

```python
classif2 = BaggingClassifier(base_estimator=classifier, n_estimators=5, random_state=1)
```

```python
classif2.fit(x_train,y_train)
```

```python
classif2.score(x_test,y_test)
```

```python
y_predict2 = classif2.predict(x_test)
```

```python
plot_learning_curves(x_train, y_train, x_test, y_test, classif2, scoring="misclassification error")
```

```python
from scipy.stats import t
from scipy import stats
accuracy1 =[0.71512571, 0.71669071, 0.71564821, 0.71570571, 0.715685, 0.71609679, 0.71509679, 0.71525929, 0.71530429]
accuracy2 =[0.76537809, 0.76594559, 0.76682558, 0.76504059, 0.7664475, 0.76637,0.76560191, 0.76490191, 0.76672192, 0.76622192]
error1 = np.subtract(1,accuracy1) #mean_accuracy of classifier 1
error2 = np.subtract(1,accuracy2) #mean_accuracy of classifier 2
mean_error1, mean_error2 = np.mean(error1), np.mean(error2)
n1,n2 = len(error1), len(error2)
df = n1+n2-2          # nfold = 10. hence,df = n1+n2-2
Significance = 0.05
critical_val = t.ppf(1.0-Significance,df)
t_stat, p=stats.ttest_ind(error1, error2, axis=0, equal_var = True)
print('Values of t_stat:',t_stat,', Critical Value:',critical_val,'and p =',p)
```

```python
if abs(t_stat) <= critical_val:
    print('Null Hypothesis is ACCEPTED that is the means are equal.')
else:
    print('Null Hypothesis is REJECTED that is the means are not equal.')
```

```python
#interpreting via p-value
#Null Hypothesis is taken as the Means of accuracy in both the classifiers
#need to be equal.Then Null Hypothesis will be true. 
if p > Significance:
    print('Null Hypothesis is ACCEPTED that is the means are equal.')
else:
    print('Null Hypothesis is REJECTED that is the means are not equal.')
```
